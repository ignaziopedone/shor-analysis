```python
import Shor_library
from qiskit import IBMQ
import random
from math import gcd
import math
```

```python
PRIMES=[3,5,7,11,13,17,23]
IBMQ.load_account();
```

```python
print('Tests using 100 shots')
for p in PRIMES:
    for q in PRIMES:
        if p<q:
            N=p*q
            print(N)
            if N<512:
                print('    Normal QFT, Basic')
                if N<64:
                    T_vec=[]
                    P_vec=[]
                    a_vec=[]
                    for a in range(2,N-1):
                        if gcd(a,N)==1:
                            (p_prime,q_prime,T,P)=Shor_library.Shor(N,a,100,sequential='False',runtime='True',prob='True')
                            T_vec.append(T)
                            P_vec.append(P)
                            a_vec.append(a)
                    print(a_vec)
                    print(T_vec)
                    print(P_vec)
                    print('----------------------------------')
                    T_vec=[]
                    P_vec=[]
                    a_vec=[]
                    print('    Normal QFT, InPlace')
                    for a in range(2,N-1):
                        if gcd(a,N)==1:
                            (p_prime,q_prime,T,P)=Shor_library.Shor(N,a,100,sequential='False',method='inplace',runtime='True',prob='True')
                            T_vec.append(T)
                            P_vec.append(P)
                            a_vec.append(a)
                    print(a_vec)
                    print(T_vec)
                    print(P_vec)
                    print('----------------------------------')
                T_vec=[]
                P_vec=[]
                a_vec=[]
                print('    Sequential QFT, Basic')
                for a in range(2,N-1):
                    if int(math.sqrt(a))==(math.sqrt(a)):
                        if gcd(a,N)==1:
                            (p_prime,q_prime,T,P)=Shor_library.Shor(N,a,100,runtime='True',prob='True')
                            T_vec.append(T)
                            P_vec.append(P)
                            a_vec.append(a)
                            print('a:',a,'P:',P)
                print(a_vec)
                print(T_vec)
                print(P_vec)
                print('----------------------------------')
                T_vec=[]
                P_vec=[]
                a_vec=[]
                print('    Sequential QFT, InPlace')
                for a in range(2,N-1):
                    if gcd(a,N)==1:
                        (p_prime,q_prime,T,P)=Shor_library.Shor(N,a,100,method='inplace',runtime='True',prob='True')
                        T_vec.append(T)
                        P_vec.append(P)
                        a_vec.append(a)
                print(a_vec)
                print(T_vec)
                print(P_vec)
                print('--------------------------------------------------------')

print('Tests using 2000 shots')
for p in PRIMES:
    for q in PRIMES:
        if p<q:
            N=p*q
            print(N)
            if N<64:
                T_vec=[]
                P_vec=[]
                a_vec=[]
                print('    Sequential QFT, Basic')
                for a in range(2,N-1):
                    if gcd(a,N)==1:
                        (p_prime,q_prime,T,P)=Shor_library.Shor(N,a,2000,runtime='True',prob='True')
                        T_vec.append(T)
                        P_vec.append(P)
                        a_vec.append(a)
                print(a_vec)
                print(T_vec)
                print(P_vec)
                print('----------------------------------')
                T_vec=[]
                P_vec=[]
                a_vec=[]
                print('    Sequential QFT, InPlace')
                for a in range(2,N-1):
                    if gcd(a,N)==1:
                        (p_prime,q_prime,T,P)=Shor_library.Shor(N,a,2000,method='inplace',runtime='True',prob='True')
                        T_vec.append(T)
                        P_vec.append(P)
                        a_vec.append(a)
                print(a_vec)
                print(T_vec)
                print(P_vec)
                print('--------------------------------------------------------')
```
